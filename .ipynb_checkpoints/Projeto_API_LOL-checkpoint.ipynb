```python
#Importando Bibliotecas
import requests
```

```python
# Recebe Links e tranforma dados em texto
link_versao = requests.get('https://ddragon.leagueoflegends.com/api/versions.json')
versao = link_versao.text
versao = list(versao.replace("[", "").replace('"', "").replace(']', "").split(","))
versao_validada = []
```

```python
for i in versao:
    campeao_string = "https://ddragon.leagueoflegends.com/cdn/" + i + "/data/pt_BR/champion.json"
    if requests.get(campeao_string).status_code == 200:
        versao_validada.append(i)
```
